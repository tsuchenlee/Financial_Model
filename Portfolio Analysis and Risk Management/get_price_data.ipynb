# Get data from yahoo finance and write in excel

```python
import pandas as pd
import numpy as np
import datetime as dt
import yfinance as yf
from pandas_datareader import data as pdr
```

```python
# stock symbol in the list
stocklist = [
    '^GSPC', 'AAPL', 'MSFT', 'AMZN', 'GOOGL', 'BRK.B', 'NVDA', 'TSLA', 'GOOG', 'XOM', 'UNH',
    'JNJ', 'JPM', 'META', 'V', 'PG', 'MA', 'CVX', 'MRK', 'LLY', 'BNPQY', 'MS', 'BARC',
    'ABBV', 'BAC', 'AVGO', 'PEP', 'KO', 'TMO', 'COST', 'DIS', 'WMT', 'CSGN', 'HSBA',
    'CSCO', 'MCD', 'ABT', 'WFC', 'TSM', 'ADBE', 'DHR', 'CRM', 'CMCSA', 'UBSG',
    'NFLX', 'TXN', 'PM', 'BMY', 'C', 'NEE', 'QCOM', 'RTX', 'COP', 'SCGLY', 'CS'
]
```

```python
len(stocklist)
```

```python
# Set the start and end date
start_date = "2013-01-01"
end_date = "2022-12-31"
```

```python
# Save price the in dataframe format
stock_data = pd.DataFrame()

for symbol in stocklist:
    stock = yf.download(symbol, start=start_date, end=end_date)
    stock['symbol'] = symbol
    stock_data = stock_data.append(stock)

# Convert the datetime index to be timezone unaware
stock_data.index = stock_data.index.tz_localize(None)

# Reset the index to have a 'Date' column
stock_data.reset_index(inplace=True)

# Pivot the data to have 'symbol' as the index and 'Date' as the columns
pivot_data = stock_data.pivot(index='symbol', columns='Date', values='Close')

# Format the 'Date' columns to have the desired format of '%Y%m%d'
pivot_data.columns = pivot_data.columns.strftime('%Y%m%d')

# Use 'with' to wrtie data to an Excel file
filename = 'stock_price_data.xlsx'
with pd.ExcelWriter(filename) as writer:
    pivot_data.to_excel(writer, index=True, sheet_name='Sheet1')
```
